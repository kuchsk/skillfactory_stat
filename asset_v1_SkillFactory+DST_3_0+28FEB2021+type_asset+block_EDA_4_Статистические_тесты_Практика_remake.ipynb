## 4. Очистка данных ##
###Проверим какие столбцы являются неинформативными и их можно удалить, обычно это столбцы
 * *С высокой долей пропусков*
 * *С низким числом уникальных значений*
 * *С высоким числом одинаковых значений*

```python
# Создаём список удалённых столбцов для отчёта
removed_columns = []

# 1. Удаление столбцов с высокой долей пропусков
threshold = 0.8
null_rates = data.isnull().mean()
cols_with_high_nulls = null_rates[null_rates >= threshold].index.tolist()
#data.drop(cols_with_high_nulls, axis=1, inplace=True)
removed_columns.extend([(col, 'высокий процент пропусков') for col in cols_with_high_nulls])

# 2. Удаление столбцов с низким числом уникальных значений
min_unique = 2
for col in data.columns:
    if data[col].nunique() <= min_unique:
        #data.drop(col, axis=1, inplace=True)
        removed_columns.append((col, 'низкое число уникальных значений'))

# 3. Удаление столбцов, где большинство значений одинаковое (>95%)
for col in data.columns:
    top_freq = data[col].value_counts(normalize=True).iloc[0]
    if top_freq > 0.95:
        #data.drop(col, axis=1, inplace=True)
        removed_columns.append((col, 'большинство значений одинаковое (>95%)'))

# Вывод отчёта
print(f"Может быть удалено столбцов: {len(removed_columns)}")
for col, reason in removed_columns:
    print(f"Столбец '{col}' может быть удалён по причине: {reason}")
```

```python
print('Проверка нормальности распределения зарплат:')
# Фильтрация данных по должности
data_ds = data[data['job_title'] == 'Data Scientist']
# Статистика
stat, p_value = stats.shapiro(data_ds['salary_in_usd'])
print(f'Статистика: {stat:.4f}, p-значение: {p_value:.4f}')

# Вывод результата проверки нормальности
if p_value < 0.05:
    print("Распределение зарплат НЕ является нормальным (p < 0.05). Используем непараметрический тест.")
else:
    print("Распределение зарплат является примерно нормальным (p >= 0.05). Можно использовать параметры тесты.")
```

### Проверка различий в уровнях зарплат по годам
Для оценки наличия статистически значимых различий между зарплатами в разные годы использован непараметрический тест Краскела-Уоллиса, который не требует предположения о нормальности распределения.

```python
# Разделение данных по годам
sal_2020 = data_ds[data_ds['work_year'] == 2020]['salary_in_usd']
sal_2021 = data_ds[data_ds['work_year'] == 2021]['salary_in_usd']
sal_2022 = data_ds[data_ds['work_year'] == 2022]['salary_in_usd']

# Тест Краскела-Уоллиса
stat, p = kruskal(sal_2020, sal_2021, sal_2022)
print(f'Статистика: {stat:.4f}, p-значение: {p:.4f}')

# Вычисляем средние или медианы по группам для определения направления изменений
mean_2020 = sal_2020.median()
mean_2021 = sal_2021.median()
mean_2022 = sal_2022.median()

# Визуализации распределений зарплат по годам
plt.figure(figsize=(10,6))
sns.boxplot(x='work_year', y='salary_in_usd', data=data_ds)
plt.title('Распределение зарплат по годам (2020-2022)')
plt.xlabel('Год')
plt.ylabel('Зарплата в USD')
plt.show()

plt.figure(figsize=(10,6))
sns.violinplot(x='work_year', y='salary_in_usd', data=data_ds)
plt.title('Распределение зарплат по годам (2020-2022)')
plt.xlabel('Год')
plt.ylabel('Зарплата в USD')
plt.show()

# Вывод результата в зависимости от p-значения и направления изменения
if p < 0.05:
    # Определяем направление изменения зарплат по медианам
    if mean_2022 > mean_2020:
        print("Статистически значимое увеличение зарплат с 2020 по 2022 годы.")
    elif mean_2022 < mean_2020:
        print("Статистически значимое снижение зарплат с 2020 по 2022 годы.")
    else:
        print("Статистически значимые изменения в зарплатах не обнаружены.")
else:
    print("Нет статистически значимых различий в зарплатах между годами.")
```

 * H0 (нулевая гипотеза): Зарплаты у специалистов Data Scientist не различаются по годам.
 * H1 (альтернативная гипотеза): Зарплаты различаются по годам.

Результат теста показал p-значение, меньшее 0.05, что свидетельствует о статистической значимости различий в уровнях зарплат между 2020, 2021 и 2022 годами.

```python
# 1. Фильтрация данных
ds_2022 = data[(data['job_title'] == 'Data Scientist') & (data['work_year'] == 2022)]
de_2022 = data[(data['job_title'] == 'Data Engineer') & (data['work_year'] == 2022)]

# 2. Визуализация распределения зарплат
plt.figure(figsize=(10, 6))
sns.boxplot(x='job_title', y='salary_in_usd', data=pd.concat([ds_2022, de_2022]))
plt.title('Сравнение зарплат Data Scientist и Data Engineer в 2022 году')
plt.xlabel('Должность')
plt.ylabel('Зарплата в USD')
plt.grid(True)
plt.show()

# 3. Расчет средних значений
mean_ds = ds_2022['salary_in_usd'].mean()
mean_de = de_2022['salary_in_usd'].mean()
print(f"Средняя зарплата Data Scientist в 2022: {mean_ds:.2f} USD")
print(f"Средняя зарплата Data Engineer в 2022: {mean_de:.2f} USD")
print(f"Разница: {(mean_ds - mean_de):.2f} USD (+{((mean_ds / mean_de) - 1)*100:.1f}%)")

# 4. Проверка нормальности распределения (тест Шапиро-Уилка)
print("\nПроверка нормальности распределения зарплат:")
stat_ds, p_ds = shapiro(ds_2022['salary_in_usd'])
stat_de, p_de = shapiro(de_2022['salary_in_usd'])
print(f"Data Scientist: p = {p_ds:.4f} {'(нормальное)' if p_ds > 0.05 else '(не нормальное)'}")
print(f"Data Engineer: p = {p_de:.4f} {'(нормальное)' if p_de > 0.05 else '(не нормальное)'}")

# 5. Выбор статистического теста
if p_ds > 0.05 and p_de > 0.05:
    # T-тест для нормальных данных
    t_stat, p_value = ttest_ind(ds_2022['salary_in_usd'], de_2022['salary_in_usd'], equal_var=False)
    test_name = "T-тест"
else:
    # Тест Манна-Уитни для ненормальных данных
    t_stat, p_value = mannwhitneyu(ds_2022['salary_in_usd'], de_2022['salary_in_usd'], alternative='two-sided')
    test_name = "Тест Манна-Уитни"

print(f"\n{test_name}")
print(f"Количество Data Scientist: {len(ds_2022)}, Data Engineer: {len(de_2022)}")
print(f"Статистика: {t_stat:.4f}, p-значение: {p_value:.4f}")

# Интерпретация результатов
if p_value > 0.05:
    print("Вывод: статистически значимых различий в зарплатах между Data Scientist и Data Engineer в 2022 году нет (не отвергаем нулевую гипотезу).")
else:
    print("Вывод: есть статистически значимые различия в зарплатах между Data Scientist и Data Engineer в 2022 году (отвергаем нулевую гипотезу).")
```

### 5.6.3 Растет ли зарплата каждый год?

```python
# Отфильтруем данные по годам
df_2020 = data[data['work_year'] == 2020]['salary_in_usd']
df_2021 = data[data['work_year'] == 2021]['salary_in_usd']
df_2022 = data[data['work_year'] == 2022]['salary_in_usd']

# Визуализация распределения зарплат по годам
plt.figure(figsize=(10,6))
sns.boxplot(x='work_year', y='salary_in_usd', data=data)
plt.title('Распределение зарплат по годам (2020-2022)')
plt.xlabel('Год')
plt.ylabel('Зарплата в USD')
plt.show()

plt.figure(figsize=(10,6))
sns.violinplot(x='work_year', y='salary_in_usd', data=data)
plt.title('Распределение зарплат по годам (2020-2022)')
plt.xlabel('Год')
plt.ylabel('Зарплата в USD')
plt.show()

# Проверка нормальности с помощью теста Шапиро-Уилка
p_values = []
for year, group in zip([2020, 2021, 2022], [df_2020, df_2021, df_2022]):
    stat, p = stats.shapiro(group)
    p_values.append(p)
    print(f"Год {year} | Статистика Шапиро-Уилка={stat:.3f}, p-value={p:.3f}")
    if p < 0.05:
        print(f"Распределение зарплат в {year} не является нормальным.")
    else:
        print(f"Распределение зарплат в {year} можно считать нормальным.")

# Выбор подходящего теста (потом сокращу)
if all(p > 0.05 for p in p_values):
    # Все группы нормальны и дисперсии равны - используем ANOVA
    F_stat, p_value = stats.f_oneway(df_2020, df_2021, df_2022)
    print(f"\nВсе группы нормальны и дисперсии равны - используем ANOVA | F-статистика={F_stat:.3f}, p-value={p_value:.3f}")
    if p_value < 0.05:
        print("Есть статистическая разница между группами зарплат по годам.")
    else:
        print("Нет статистической разницы между группами зарплат по годам.")
else:
    # Не все условия выполнены — используем Краскел-Уоллиса
    h_statistic, p_value = stats.kruskal(df_2020, df_2021, df_2022)
    print(f"\nНе все условия выполнены — используем Краскел-Уоллиса | H-статистика={h_statistic:.3f}, p-value={p_value:.3f}")
    if p_value < 0.05:
        print("Есть статистическая разница между группами зарплат по годам.")
    else:
        print("Нет статистической разницы между группами зарплат по годам.")
```

### 5.6.4 Проверим как соотносятся уровень опыта с удаленной работой

```python
# Создаем кросс-таблицу: experience_level vs employment_type
ct_experience_employment = pd.crosstab(data['experience_level'], data['remote_ratio'])

# Визуализация: столбчатая диаграмма для experience_level по employment_type
ct_experience_employment.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Опыт работы vs Удаленный коэффициент')
plt.xlabel('Уровень опыта')
plt.ylabel('Удаленный коэффициент')
plt.legend(title='Удаленка')
plt.show()

# Создаем таблицу сопряженности (кросс-таблицу)
ct = pd.crosstab(data['experience_level'], data['remote_ratio'])

print("Кросс-таблица:\n", ct)

# Выполняем χ²-тест независимости
chi2, p, dof, ex = chi2_contingency(ct)

print(f'χ²-статистика: {chi2}')
print(f'p-value: {p}')

# Интерпретация
alpha = 0.05
if p < alpha:
    print("Есть статистически значимая зависимость между уровнем опыта и удаленным коэффициентом (отклоняем H0).")
else:
    print("Нет статистически значимой зависимости между уровнем опыта и удаленным коэффициентом (не отвергаем H0).")
```

### 5.6.7 Выясним зависимость между средней зарплатой и количеством вакансий в странах

```python
# Группируем по странам и считаем среднюю зарплату и количество вакансий
grouped = data.groupby('company_location').agg(average_salary=('salary_in_usd', 'mean'), vacancy_count=('salary_in_usd', 'size')).reset_index()

# Получаем топ-20 стран по средней зарплате
top20 = grouped.sort_values('average_salary', ascending=False).head(20)

# Проверим нормальность для средней зарплаты
stat_salary, p_salary = shapiro(top20['average_salary'])
print(f"Нормальность средней зарплаты: p-value={p_salary:.3f}")

# Проверим нормальность для количества вакансий
stat_vacancy, p_vacancy = shapiro(top20['vacancy_count'])
print(f"Нормальность количества вакансий: p-value={p_vacancy:.3f}")

# Визуализация
sns.set(style="darkgrid")
fig, ax1 = plt.subplots(figsize=(12, 8))

# Барплот для средней зарплаты
sns.barplot(x='average_salary', y='company_location', data=top20, ax=ax1)
ax1.set_xlabel('Средняя зарплата в USD')
ax1.set_ylabel('Страна компании')
ax1.set_title('Топ 20 стран по средней зарплате и количеству вакансий')

# Создаем вторую ось для количества вакансий
ax2 = ax1.twiny()

# Строим точки для количества вакансий и добавляем интерпретацию
sns.scatterplot(x='vacancy_count', y='company_location', data=top20, ax=ax2, color='orange', s=100, label='Количество вакансий')

ax2.set_xlabel('Количество вакансий')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()

# Проверим нормальность для средней зарплаты
stat_salary, p_salary = shapiro(top20['average_salary'])
print(f"Нормальность средней зарплаты: p-value={p_salary:.3f}")

# Проверим нормальность для количества вакансий
stat_vacancy, p_vacancy = shapiro(top20['vacancy_count'])
print(f"Нормальность количества вакансий: p-value={p_vacancy:.3f}")

# Расчет корреляции Спирмена
correlation_spearman = top20['average_salary'].corr(top20['vacancy_count'], method='spearman')
print(f"Коэффициент корреляции Спирмена: {correlation_spearman:.2f}")

# Интерпретация
if correlation_spearman > 0.7:
    interpretation = "сильной положительной связи"
elif correlation_spearman > 0.4:
    interpretation = "умеренной положительной связи"
elif correlation_spearman > 0.1:
    interpretation = "слабой положительной связи"
elif correlation_spearman < -0.1:
    interpretation = "слабой отрицательной связи"
elif correlation_spearman < -0.4:
    interpretation = "умеренной отрицательной связи"
else:
    interpretation = "отсутствии значимой связи"

print(f"Это говорит о {interpretation} между средней зарплатой и количеством вакансий по странам.")
```

```python
!pip install -q dash
!pip install -q pycountry

import dash
from dash import Dash, dcc, html, Input, Output, State
import plotly.express as px
import pycountry

# Преобразование кодов стран в имена
iso2_to_name = {country.alpha_2: country.name for country in pycountry.countries}
data['country_name'] = data['company_location'].map(iso2_to_name)

# Уникальные значения для фильтров
def get_options(series):
    return [{'label': str(val), 'value': val} for val in sorted(series.unique())]

# Создаем словарь с возможными фильтрами
filters_config = {
    'year': {
        'label': 'Год',
        'options': get_options(data['work_year']),
        'initial': [str(year) for year in data['work_year'].unique()]
    },
    'experience': {
        'label': 'Уровень опыта',
        'options': get_options(data['experience_level']),
        'initial': list(data['experience_level'].unique())
    },
    'employment': {
        'label': 'Тип занятости',
        'options': get_options(data['employment_type']),
        'initial': list(data['employment_type'].unique())
    },
    'job': {
        'label': 'Должность',
        'options': get_options(data['job_title']),
        'initial': list(data['job_title'].unique())
    }
}

# Универсальная функция для создания групп фильтров с кнопками "Все" и "Снять"
def create_filter_group(filter_id, label, options, initial):
    return html.Div([
        html.H5(label),
        html.Button('Все', id=f'select-all-{filter_id}', n_clicks=0),
        html.Button('Снять', id=f'deselect-all-{filter_id}', n_clicks=0),
        dcc.Checklist(
            id=filter_id,
            options=options,
            value=initial,
            inline=True,
            labelStyle={'margin-right': '15px'}
        )
    ], style={'padding': '10px', 'border': '1px solid #ddd', 'margin': '10px', 'background': '#f9f9f9'})

# Создаем элементы фильтров
filter_groups = []
for key, config in filters_config.items():
    filter_groups.append(
        create_filter_group(key, config['label'], config['options'], config['initial'])
    )

# Основной лейаут
app = Dash(__name__)
app.layout = html.Div([
    html.H1("Интерактивная карта зарплат", style={'textAlign': 'center'}),
    html.Div(filter_groups, style={'display': 'flex', 'flex-wrap': 'wrap', 'justify-content': 'space-around'}),
    dcc.Graph(id='world-map', style={'height': '80vh'})
])

# Обработка кнопок "Все"/"Снять" для каждого фильтра
for key in filters_config.keys():
    @app.callback(
        Output(key, 'value'),
        [Input(f'select-all-{key}', 'n_clicks'),
         Input(f'deselect-all-{key}', 'n_clicks')],
        State(key, 'options')
    )
    def update_checklist(select_all_clicks, deselect_all_clicks, options, filter_id=key):
        ctx = dash.callback_context
        if not ctx.triggered:
            return [opt['value'] for opt in options]
        button_id = ctx.triggered[0]['prop_id'].split('.')[0]
        if button_id == f'select-all-{filter_id}':
            return [opt['value'] for opt in options]
        elif button_id == f'deselect-all-{filter_id}':
            return []
        return [opt['value'] for opt in options]

# Обновляем карту при изменении фильтров
@app.callback(
    Output('world-map', 'figure'),
    [Input(f'{key}', 'value') for key in filters_config.keys()]
)
def update_map(years, exp_levels, emp_types, jobs):
    # Преобразуем годы в строки для сравнения
    years_str = [str(y) for y in years]

    filtered = data[
        (data['work_year'].astype(str).isin(years_str)) &
        (data['experience_level'].isin(exp_levels)) &
        (data['employment_type'].isin(emp_types)) &
        (data['job_title'].isin(jobs))
    ]

    if filtered.empty:
        # Пустая карта, если данных нет
        fig = px.choropleth()
        fig.update_layout(title='Нет данных по выбранным фильтрам')
        return fig

    # Аггрегация по странам
    aggregated = filtered.groupby('country_name', as_index=False)['salary_in_usd'].mean()

    # Вычисление корреляции
    if len(aggregated) > 1:
        # Можно использовать корреляцию Спирмена, учитывая распределение
        correlation = aggregated['salary_in_usd'].corr(filtered['salary_in_usd'], method='spearman')
        # Интерпретация
        if correlation > 0.7:
            interpretation = "Сильная положительная связь"
        elif correlation > 0.4:
            interpretation = "Умеренная положительная связь"
        elif correlation > 0.1:
            interpretation = "Слабая положительная связь"
        elif correlation < -0.1:
            interpretation = "Слабая отрицательная связь"
        elif correlation < -0.4:
            interpretation = "Умеренная отрицательная связь"
        elif correlation < -0.7:
            interpretation = "Сильная отрицательная связь"
        else:
            interpretation = "Отсутствие значимой связи"
    else:
        correlation = None
        interpretation = "Недостаточно данных для оценки связи"

    # Построение карты
    fig = px.choropleth(
        aggregated,
        locations='country_name',
        locationmode='country names',
        color='salary_in_usd',
        hover_name='country_name',
        color_continuous_scale='Viridis',
        labels={'salary_in_usd': 'Зарплата (USD)'},
        title='Средние зарплаты по странам'
    )

    # Добавляем подпись с интерпретацией
    annotation_text = f"Коэффициент корреляции (Спирмен): {correlation:.2f}<br>{interpretation}"

    fig.add_annotation(
        text=annotation_text,
        align='left',
        showarrow=False,
        xref='paper',
        yref='paper',
        x=0,
        y=-0.2,
        bordercolor='black',
        borderwidth=1,
        bgcolor='white'
    )

    # Настройка формы и оформления
    fig.update_layout(
        coloraxis_colorbar=dict(title='USD', x=0.85),
        margin=dict(l=0, r=0, t=40, b=80),
        title=dict(text='Средние зарплаты по странам', x=0.5, font=dict(size=20))
    )

    fig.update_traces(
        hovertemplate="<b>%{hovertext}</b><br>Зарплата: %{z:,.0f} USD"
    )

    return fig


if __name__ == '__main__':
    app.run(debug=True)
    # app.run_server(debug=True)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.stats.outliers_influence import variance_inflation_factor
import plotly.express as px
import textwrap

def ask_binary(question, hint):
    wrapped_question = textwrap.fill(question, width=80)
    wrapped_hint = textwrap.fill(hint, width=80)
    while True:
        print(f"\n{wrapped_question}\nПодсказка: {wrapped_hint}\n(1 — да, 0 — нет):")
        ans = input().strip()
        if ans in ('0', '1'):
            return ans == '1'
        print("Ошибка: введите 1 или 0.")

def ask_choice(question, choices, hint):
    wrapped_question = textwrap.fill(question, width=80)
    wrapped_hint = textwrap.fill(hint, width=80)
    choices_str = '/'.join(choices)
    while True:
        print(f"\n{wrapped_question}\nВарианты: {choices_str}\nПодсказка: {wrapped_hint}")
        ans = input("Введите выбор: ").strip()
        if ans in choices:
            return ans
        print(f"Ошибка: выберите один из вариантов: {choices_str}")

def print_summary(data):
    print("\n--- Общая информация о данных ---")
    data.info()
    print("\nСтатистические показатели для числовых данных:")
    print(data.describe(include=[np.number]))
    print("\nУникальные значения для категориальных данных:")
    print(data.describe(include=[object]))
    print("\nПроцент пропусков по столбцам:")
    print((data.isnull().sum() / len(data) * 100).round(2))

    numeric_cols = data.select_dtypes(include=[np.number]).columns
    if len(numeric_cols) > 0:
        plt.figure(figsize=(12, 6))
        for col in numeric_cols:
            sns.histplot(data[col].dropna(), kde=True, label=col)
        plt.title("Гистограммы числовых столбцов")
        plt.legend()
        plt.show()

        if len(numeric_cols) > 1:
            sns.pairplot(data[numeric_cols])
            plt.title("Парные графики числовых столбцов")
            plt.show()

    categorical_cols = data.select_dtypes(include=[object]).columns
    for col in categorical_cols:
        plt.figure(figsize=(8, 6))
        data[col].value_counts().plot(kind='pie', autopct='%1.1f%%')
        plt.title(f"Распределение '{col}'")
        plt.ylabel('')
        plt.show()

def is_index_like_column(data, col, threshold=0.95):
    if data[col].nunique() / len(data) >= threshold:
        if np.issubdtype(data[col].dtype, np.number):
            diff = np.diff(data[col].dropna())
            if len(diff) > 0 and np.all(diff == diff[0]):
                return True
        return True
    return False

def detect_duplicate_columns(data, corr_threshold=0.95, log_file=None):
    to_drop = []
    numeric_cols = data.select_dtypes(include=[np.number]).columns
    categorical_cols = data.select_dtypes(include=[object]).columns

    if len(numeric_cols) > 1:
        corr_matrix = data[numeric_cols].corr()
        plt.figure(figsize=(10, 8))
        sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
        plt.title("Корреляционная матрица числовых столбцов")
        plt.show()

        for i in range(len(numeric_cols)):
            for j in range(i + 1, len(numeric_cols)):
                if abs(corr_matrix.iloc[i, j]) >= corr_threshold:
                    col1, col2 = numeric_cols[i], numeric_cols[j]
                    print(f"Столбцы '{col1}' и '{col2}' имеют корреляцию {corr_matrix.iloc[i, j]:.3f}.")
                    if ask_binary(f"Удалить столбец '{col2}'?", f"Оставить '{col1}' (высокая корреляция может быть критичной)"):
                        to_drop.append(col2)
                        if log_file:
                            with open(log_file, 'a') as f:
                                f.write(f"Удалён столбец '{col2}' (высокая корреляция с '{col1}': {corr_matrix.iloc[i, j]:.3f})\n")

    for i in range(len(categorical_cols)):
        for j in range(i + 1, len(categorical_cols)):
            col1, col2 = categorical_cols[i], categorical_cols[j]
            similarity = (data[col1] == data[col2]).mean()
            if similarity >= corr_threshold:
                print(f"Столбцы '{col1}' и '{col2}' имеют {similarity:.2%} совпадающих значений.")
                if ask_binary(f"Удалить столбец '{col2}'?", f"Оставить '{col1}' (высокая схожесть может быть критичной)"):
                    to_drop.append(col2)
                    if log_file:
                        with open(log_file, 'a') as f:
                            f.write(f"Удалён столбец '{col2}' (схожесть с '{col1}': {similarity:.2%})\n")

    if to_drop:
        data = data.drop(columns=to_drop)
        print(f"Удалены дублирующиеся столбцы: {to_drop}")
    return data

def calculate_vif(data, numeric_cols, log_file=None):
    if len(numeric_cols) < 2:
        return []
    X = data[numeric_cols].dropna()
    vif_data = pd.DataFrame()
    vif_data["Variable"] = numeric_cols
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    print("\n--- Проверка мультиколлинеарности (VIF) ---")
    print(vif_data)
    plt.figure(figsize=(8, 6))
    sns.barplot(x="VIF", y="Variable", data=vif_data)
    plt.title("VIF для числовых столбцов")
    plt.show()
    to_drop = []
    for _, row in vif_data[vif_data["VIF"] > 10].iterrows():
        col = row["Variable"]
        print(f"Столбец '{col}' имеет высокий VIF ({row['VIF']:.2f}).")
        if ask_binary(f"Удалить столбец '{col}'?", "Высокий VIF может быть критичным"):
            to_drop.append(col)
            if log_file:
                with open(log_file, 'a') as f:
                    f.write(f"Удалён столбец '{col}' (высокий VIF: {row['VIF']:.2f})\n")
    return to_drop

def analyze_columns(data, unique_threshold=0.1, missing_threshold=0.8, log_file=None):
    to_drop = []
    print("\n--- Анализ столбцов на информативность ---")
    for col in data.columns:
        unique_ratio = data[col].nunique(dropna=True) / len(data)
        missing_ratio = data[col].isnull().sum() / len(data)

        if is_index_like_column(data, col):
            print(f"Столбец '{col}' похож на индекс (уникальность: {unique_ratio:.2%}). Автоматическое удаление.")
            to_drop.append(col)
            if log_file:
                with open(log_file, 'a') as f:
                    f.write(f"Удалён столбец '{col}' (индексоподобный, уникальность: {unique_ratio:.2%})\n")
            continue

        if unique_ratio <= unique_threshold and data[col].nunique(dropna=False) > 5:
            print(f"Столбец '{col}' имеет низкую уникальность ({unique_ratio:.2%}).")
            if ask_binary(f"Удалить столбец '{col}'?", "Мало уникальных значений, но может быть важен"):
                to_drop.append(col)
                if log_file:
                    with open(log_file, 'a') as f:
                        f.write(f"Удалён столбец '{col}' (низкая уникальность: {unique_ratio:.2%})\n")
                continue

        if missing_ratio >= missing_threshold:
            print(f"Столбец '{col}' содержит {missing_ratio:.2%} пропусков.")
            if ask_binary(f"Удалить столбец '{col}'?", "Слишком много пропусков, но может быть важен"):
                to_drop.append(col)
                if log_file:
                    with open(log_file, 'a') as f:
                        f.write(f"Удалён столбец '{col}' (пропуски: {missing_ratio:.2%})\n")
                continue

    numeric_cols = data.select_dtypes(include=[np.number]).columns
    if len(numeric_cols) > 1:
        vif_drops = calculate_vif(data, numeric_cols, log_file)
        to_drop.extend(vif_drops)

    if to_drop:
        data = data.drop(columns=to_drop)
        print(f"Удалены столбцы: {to_drop}")

    plt.figure(figsize=(10, 6))
    sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
    plt.title("Карта пропусков в данных")
    plt.show()

    return data

def detect_outliers(data, col, log_file=None):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = data[col][(data[col] < Q1 - 1.5 * IQR) | (data[col] > Q3 + 1.5 * IQR)]
    if not outliers.empty:
        print(f"Обнаружены выбросы в '{col}': {len(outliers)} значений")
        if ask_binary(f"Удалить выбросы в '{col}'?", "На основе IQR"):
            data = data[~data[col].isin(outliers)]
            if log_file:
                with open(log_file, 'a') as f:
                    f.write(f"Удалены выбросы в '{col}': {len(outliers)} значений\n")
    return data

def remove_duplicates(data, log_file=None):
    if ask_binary("Удалить дублирующиеся строки?", "Проверяет полное совпадение строк"):
        initial_rows = len(data)
        data = data.drop_duplicates()
        duplicates_removed = initial_rows - len(data)
        print(f"Удалено {duplicates_removed} дублирующихся строк.")
        if log_file:
            with open(log_file, 'a') as f:
                f.write(f"Удалено {duplicates_removed} дублирующихся строк\n")
    return data

def handle_missing(data, log_file=None):
    for col in data.columns:
        if data[col].isnull().sum() > 0:
            dtype = data[col].dtype
            if np.issubdtype(dtype, np.number):
                choice = ask_choice(
                    f"Заполнить пропуски в столбце '{col}'?",
                    ['медиана', 'среднее', 'пропуск'],
                    "Выберите способ заполнения пропусков"
                )
                if choice == 'медиана':
                    data[col].fillna(data[col].median(), inplace=True)
                    if log_file:
                        with open(log_file, 'a') as f:
                            f.write(f"Пропуски в '{col}' заполнены медианой\n")
                elif choice == 'среднее':
                    data[col].fillna(data[col].mean(), inplace=True)
                    if log_file:
                        with open(log_file, 'a') as f:
                            f.write(f"Пропуски в '{col}' заполнены средним\n")
                else:
                    print(f"Пропуски в '{col}' оставлены без изменений.")
            else:
                choice = ask_choice(
                    f"Заполнить пропуски в столбце '{col}'?",
                    ['мода', 'пропуск'],
                    "Выберите способ заполнения пропусков"
                )
                if choice == 'мода':
                    data[col].fillna(data[col].mode()[0], inplace=True)
                    if log_file:
                        with open(log_file, 'a') as f:
                            f.write(f"Пропуски в '{col}' заполнены модой\n")
                else:
                    print(f"Пропуски в '{col}' оставлены без изменений.")
    return data

def select_columns(data):
    print("\nДоступные столбцы:")
    for idx, col in enumerate(data.columns, 1):
        print(f"{idx}: {col}")
    selected_cols = []
    while True:
        inp = input("Введите номер или имя столбца для анализа (или 'стоп' для завершения): ").strip()
        if inp.lower() == 'стоп':
            break
        if inp.isdigit():
            num = int(inp)
            if 1 <= num <= len(data.columns):
                selected_cols.append(data.columns[num - 1])
            else:
                print(f"Ошибка: номер должен быть от 1 до {len(data.columns)}.")
        elif inp in data.columns:
            selected_cols.append(inp)
        else:
            print("Ошибка: неверное имя столбца.")
    return selected_cols

def analyze_numeric_columns(data, columns, log_file=None):
    for col in columns:
        print(f"\nАнализ числового столбца '{col}':")
        data = detect_outliers(data, col, log_file)
        stat, p_value = stats.shapiro(data[col].dropna())
        normal = p_value > 0.05
        print(f"Тест Шапиро-Уилка: p-value = {p_value:.3f} — распределение {'нормальное' if normal else 'ненормальное'}.")

        plt.figure(figsize=(6, 6))
        stats.probplot(data[col].dropna(), dist="norm", plot=plt)
        plt.title(f"Q-Q график для '{col}'")
        plt.show()

        unique_vals = data[col].nunique(dropna=True)
        size = data[col].dropna().shape[0]
        print(f"Количество уникальных значений: {unique_vals}")
        print(f"Размер выборки: {size}")

        if normal:
            pop_var = float(input(f"Введите гипотетическую дисперсию для '{col}' (0 для пропуска): "))
            if pop_var > 0:
                n = len(data[col].dropna())
                sample_var = data[col].var()
                chi2_stat = (n - 1) * sample_var / pop_var
                p_value = 2 * min(stats.chi2.cdf(chi2_stat, n - 1), 1 - stats.chi2.cdf(chi2_stat, n - 1))
                print(f"Тест хи-квадрат на дисперсию: chi2 = {chi2_stat:.3f}, p-value = {p_value:.3f}")
                if log_file:
                    with open(log_file, 'a') as f:
                        f.write(f"Тест хи-квадрат на дисперсию для '{col}': chi2 = {chi2_stat:.3f}, p-value = {p_value:.3f}\n")

        group_col = None
        for c in data.columns:
            if c != col and np.issubdtype(data[c].dtype, object):
                group_col = c
                break

        if group_col:
            groups = data[group_col].dropna().unique()
            print(f"Обнаружена группировка по '{group_col}': {groups}")
            group_data = [data[data[group_col] == g][col].dropna() for g in groups]
            group_sizes = [len(g) for g in group_data]
            print(f"Группы: {', '.join([f'{g} (n={s})' for g, s in zip(groups, group_sizes)])}")

            all_normal = True
            for g, g_data in zip(groups, group_data):
                stat, p = stats.shapiro(g_data)
                print(f"{g}: p={p:.3f} ({'нормальное' if p > 0.05 else 'ненормальное'})")
                if p <= 0.05:
                    all_normal = False

            if len(groups) == 2:
                group1, group2 = group_data
                paired = ask_binary("Данные зависимы (парные)?", "Например, измерения до и после лечения")
                if all_normal:
                    stat, p_levene = stats.levene(group1, group2)
                    print(f"Тест Левена на равенство дисперсий: p-value = {p_levene:.3f}")
                    if paired:
                        print("Используем парный t-тест.")
                        stat, p_value = stats.ttest_rel(group1, group2)
                        test_name = "Парный t-тест"
                    else:
                        print("Используем независимый t-тест.")
                        stat, p_value = stats.ttest_ind(group1, group2, equal_var=p_levene > 0.05)
                        test_name = "Независимый t-тест"
                else:
                    if paired:
                        print("Используем тест Вилкоксона.")
                        stat, p_value = stats.wilcoxon(group1, group2)
                        test_name = "Вилкоксона"
                    else:
                        print("Используем тест Манна-Уитни.")
                        stat, p_value = stats.mannwhitneyu(group1, group2)
                        test_name = "Манна-Уитни"
                print(f"\nРезультат {test_name}: статистика = {stat:.3f}, p-value = {p_value:.3f}")
                if log_file:
                    with open(log_file, 'a') as f:
                        f.write(f"Тест {test_name} для '{col}': статистика = {stat:.3f}, p-value = {p_value:.3f}\n")
                fig = px.box(data, x=group_col, y=col, title=f"{test_name} для '{col}'")
                fig.show()
            elif len(groups) > 2:
                paired = ask_binary("Данные зависимы (парные)?", "Например, повторные измерения на одних субъектах")
                if paired:
                    print("Используем тест Фридмана.")
                    stat, p_value = stats.friedmanchisquare(*group_data)
                    test_name = "Фридмана"
                else:
                    if all_normal:
                        stat, p_levene = stats.levene(*group_data)
                        print(f"Тест Левена: p-value = {p_levene:.3f}")
                        if p_levene > 0.05:
                            print("Используем однофакторный ANOVA.")
                            stat, p_value = stats.f_oneway(*group_data)
                            test_name = "Однофакторный ANOVA"
                            if p_value < 0.05:
                                print("Проводим пост-хок тест Тьюки...")
                                tukey = pairwise_tukeyhsd(data[col].dropna(), data[group_col].dropna())
                                print(tukey)
                                if log_file:
                                    with open(log_file, 'a') as f:
                                        f.write(f"Пост-хок тест Тьюки для '{col}':\n{tukey}\n")
                        else:
                            print("Дисперсии неоднородны, используем Краскела-Уоллиса.")
                            stat, p_value = stats.kruskal(*group_data)
                            test_name = "Краскела-Уоллиса"
                    else:
                        print("Используем Краскела-Уоллиса (ненормальные данные).")
                        stat, p_value = stats.kruskal(*group_data)
                        test_name = "Краскела-Уоллиса"
                print(f"\nРезультат {test_name}: статистика = {stat:.3f}, p-value = {p_value:.3f}")
                if log_file:
                    with open(log_file, 'a') as f:
                        f.write(f"Тест {test_name} для '{col}': статистика = {stat:.3f}, p-value = {p_value:.3f}\n")
                fig = px.box(data, x=group_col, y=col, title=f"{test_name} для '{col}'")
                fig.show()
        else:
            if size < 5:
                print("Размер выборки мал, автоматический анализ невозможен.")
                continue
            popmean = float(input(f"Введите гипотетическое среднее для '{col}': "))
            if normal:
                print("Используем одновыборочный t-тест.")
                stat, p_value = stats.ttest_1samp(data[col].dropna(), popmean)
                test_name = "Одновыборочный t-тест"
            else:
                print("Используем тест Вилкоксона.")
                stat, p_value = stats.wilcoxon(data[col].dropna() - popmean)
                test_name = "Вилкоксона"
            print(f"\nРезультат {test_name}: статистика = {stat:.3f}, p-value = {p_value:.3f}")
            if log_file:
                with open(log_file, 'a') as f:
                    f.write(f"Тест {test_name} для '{col}': статистика = {stat:.3f}, p-value = {p_value:.3f}\n")
            fig = px.box(data, y=col, title=f"{test_name} для '{col}'")
            fig.add_hline(y=popmean, line_dash="dash", line_color="red", annotation_text=f"popmean={popmean}")
            fig.show()

        other_numeric = [c for c in columns if c != col]
        if other_numeric:
            for other_col in other_numeric:
                print(f"\nАнализ корреляции между '{col}' и '{other_col}':")
                stat_other, p_other = stats.shapiro(data[other_col].dropna())
                both_normal = normal and p_other > 0.05
                if both_normal:
                    print("Используем корреляцию Пирсона.")
                    corr, p_value = stats.pearsonr(data[col].dropna(), data[other_col].dropna())
                    test_name = "Пирсона"
                else:
                    print("Используем корреляцию Кендалла.")
                    corr, p_value = stats.kendalltau(data[col].dropna(), data[other_col].dropna())
                    test_name = "Кендалла"
                print(f"Корреляция {test_name}: коэффициент = {corr:.3f}, p-value = {p_value:.3f}")
                if log_file:
                    with open(log_file, 'a') as f:
                        f.write(f"Корреляция {test_name} между '{col}' и '{other_col}': коэффициент = {corr:.3f}, p-value = {p_value:.3f}\n")
                fig = px.scatter(data, x=col, y=other_col, title=f"Корреляция {test_name} ({corr:.3f})")
                fig.show()

def analyze_categorical_columns(data, columns, log_file=None):
    for col in columns:
        print(f"\nАнализ категориального столбца '{col}':")
        other_cols = [c for c in columns if c != col]
        if not other_cols:
            print("Нет других категориальных столбцов для анализа.")
            continue
        for other_col in other_cols:
            contingency_table = pd.crosstab(data[col], data[other_col])
            print(f"\nТаблица сопряженности между '{col}' и '{other_col}':")
            print(contingency_table)
            chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
            print(f"\nКритерий Хи-квадрат: chi2 = {chi2:.3f}, p-value = {p:.3f}")
            if p < 0.05:
                print("Есть статистически значимая связь.")
            else:
                print("Связь статистически незначима.")
            if log_file:
                with open(log_file, 'a') as f:
                    f.write(f"Хи-квадрат между '{col}' и '{other_col}': chi2 = {chi2:.3f}, p-value = {p:.3f}\n")
            if contingency_table.values.min() < 5:
                print("Малые ожидаемые частоты, используем тест Фишера.")
                odds_ratio, p_value = stats.fisher_exact(contingency_table)
                print(f"Тест Фишера: p-value = {p_value:.3f}")
                if log_file:
                    with open(log_file, 'a') as f:
                        f.write(f"Тест Фишера между '{col}' и '{other_col}': p-value = {p_value:.3f}\n")
            plt.figure(figsize=(8, 6))
            sns.heatmap(contingency_table, annot=True, cmap='Blues')
            plt.title(f"Таблица сопряженности '{col}' и '{other_col}'")
            plt.show()

def main(data, log_file="data_cleaning_report.txt"):
    with open(log_file, 'w') as f:
        f.write("Отчёт по очистке и анализу данных\n")
        f.write(f"Дата и время: {pd.Timestamp.now()}\n\n")
    print_summary(data)
    data = analyze_columns(data, unique_threshold=0.1, missing_threshold=0.8, log_file=log_file)
    data = detect_duplicate_columns(data, corr_threshold=0.95, log_file=log_file)
    data = remove_duplicates(data, log_file=log_file)
    data = handle_missing(data, log_file=log_file)
    selected_cols = select_columns(data)
    numeric_cols = [c for c in selected_cols if np.issubdtype(data[c].dtype, np.number)]
    categorical_cols = [c for c in selected_cols if not np.issubdtype(data[c].dtype, np.number)]
    if numeric_cols:
        analyze_numeric_columns(data, numeric_cols, log_file=log_file)
    if categorical_cols:
        analyze_categorical_columns(data, categorical_cols, log_file=log_file)
    return data

if __name__ == '__main__':
    data = data
    # data = pd.DataFrame({
    #     'index': range(10),
    #     'const': ['A'] * 10,
    #     'year': [2020, 2020, 2020, 2021, 2021, 2021, 2022, 2022, 2022, 2022],
    #     'company_size': ['M', 'M', 'M', 'L', 'L', 'L', 'S', 'S', 'S', 'S'],
    #     'value': [10, 12, 14, 15, 20, 22, 25, 30, 28, 35],
    #     'value_copy': [10, 12, 14, 15, 20, 22, 25, 30, 28, 35],
    #     'missing': [1, None, None, None, None, None, None, None, None, 2]
    # })
    data = main(data)
```
